```python
Item_Purchased=df['Item Purchased'].value_counts().reset_index()
Item_Purchased
```

```python
plt.figure(figsize=(20, 8))
df['Item Purchased'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of Item purchased')
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
item = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()
item
```

```python
plt.figure(figsize=(12, 5))
norm = plt.Normalize(item['Purchase Amount (USD)'].min(), item['Purchase Amount (USD)'].max())
colors = plt.cm.winter(norm(item['Purchase Amount (USD)']))

# Plot using the gradient colors
sns.barplot(data=item,
            x='Item Purchased',
            y='Purchase Amount (USD)',
            palette=colors)

plt.title('Item Purchased by Amount (USD)', fontsize=14)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df['Category'].value_counts().plot(kind='bar')
plt.title('cloth category')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution by Gender')
plt.tight_layout()
```

```python
df1 = df.dropna(subset=['Age'])
```
